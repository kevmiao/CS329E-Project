```python
# Distributions of Discrete Categorical Variables
print(pd.Series(survey_df['GENDER']).mode()[0])
print(pd.Series(survey_df['AGE']).mode()[0])
print(pd.Series(survey_df['SMOKING']).mode()[0])
print(pd.Series(survey_df['YELLOW_FINGERS']).mode()[0])
print(pd.Series(survey_df['ANXIETY']).mode()[0])
print(pd.Series(survey_df['PEER_PRESSURE']).mode()[0])
print(pd.Series(survey_df['CHRONIC DISEASE']).mode()[0])
print(pd.Series(survey_df['WHEEZING']).mode()[0])
print(pd.Series(survey_df['ALCOHOL CONSUMING']).mode()[0])
print(pd.Series(survey_df['COUGHING']).mode()[0])
print(pd.Series(survey_df['SHORTNESS OF BREATH']).mode()[0])
print(pd.Series(survey_df['SWALLOWING DIFFICULTY']).mode()[0])
print(pd.Series(survey_df['LUNG_CANCER']).mode()[0])
```

```python
# Find correlation among columns, non-categorical
pd.set_option('display.max_columns', None)
correlation_matrix = survey_df.corr(numeric_only=True)
correlation_matrix
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Correlation coefficient between Coughing and Wheezing
corr_coefficient = correlation_matrix.loc['COUGHING', 'WHEEZING']
print(corr_coefficient)
```

```python
# Regression plot between Coughing and Wheezing
sns.regplot(x = 'WHEEZING', y ='COUGHING', data = survey_df)
plt.ylim(0,)
```
