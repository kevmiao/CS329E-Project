### Geographical distribution of lung cancer incidence in the US


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt
import os
```

```python
harvard_df = pd.read_csv('cleaned_air_cancer_df.csv')
```

```python
harvard_df.head()
```

```python
harvard_df.info()
```

```python
# Aggregate Lung Cancer incidence rates by State to get the average incidence rate per state
state_incidence_rates = harvard_df.groupby('State')['Lung Cancer'].mean().reset_index()

# Now proceed to create the choropleth map as you've previously attempted
fig = px.choropleth(
    state_incidence_rates,
    locations='State',  # DataFrame column with locations (state abbreviations)
    locationmode="USA-states",  # Set to use US states
    color='Lung Cancer',  # DataFrame column with color values
    color_continuous_scale='bluered',  # Color scale from blue (low) to red (high)
    scope="usa",  # Focus map on the USA
    hover_name='State',  # Column to display on hover
    hover_data={'Lung Cancer': ':.2f'},
    labels={'Lung Cancer': 'Lung Cancer Incidence Rate'}  # Label for the color bar
)

fig.update_layout(
    title_text='Geographical Distribution of Lung Cancer Incidence Rates in the US',
    title_x = 0.5,
    geo=dict(
        lakecolor='rgb(255, 255, 255)'  # Color for lakes
    ),
    coloraxis_colorbar=dict(
        title='Incidence Rate',  # Title for the color bar
        x=0.75,  # Adjust the x-position of the color bar
        y=0.45,  # Adjust the y-position of the color bar
        tickfont=dict(size=10),  # Set font size for the color bar ticks
        titlefont=dict(size=12),
        ticks='outside',  # Show ticks outside the color bar
        ticklen=5
    ),
    font = dict(
        # family="Times New Roman",  # Specify font family
        size=14,         # Specify font size
        color="black"  
    ),
    # height = 500,
    # width = 500
)

fig.add_annotation(
    text='(Number of lung cancer cases per 100,000 people in a county)',
    xref='paper', yref='paper',
    x=0.5, y=-0.1,
    showarrow=False,
    font=dict(size=10)
)

fig.add_annotation(
    text='No data',
    x=0.5, y=0.5,
    xref='paper', yref='paper',
    font=dict(size=12, color='black'),
    showarrow=True,
    arrowhead=2,
    arrowsize=1,
    arrowwidth=2,
    arrowcolor='black',
    ax=95, ay=-90  # Adjusted ax and ay values for longer arrow and top right position
)


fig.show()
```

### Sociodemographic variables and their relationship with lung cancer incidence


```python
# Print the current working directory
print(os.getcwd())

# Specify the full path to your CSV file
file_path_csv = '/Users/moosarahman/Desktop/cs329e/proj/survey_lung_cancer.csv'
# Load the dataset
df_survey = pd.read_csv(file_path_csv)

# Display the first few rows to check it's loaded correctly
print(df_survey.head())


```

```python
# Assuming 'df_survey' has the correct columns as printed above
# Calculate the correlation matrix using numeric columns only
# Note that we exclude 'GENDER' because it's categorical, and we've already encoded it as 'GENDER_NUM'

numeric_cols = ['AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE',
                'FATIGUE ', 'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING',
                'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN', 'LUNG_CANCER_NUM',
                'GENDER_NUM']  # Include 'GENDER_NUM' if you want to include gender in the correlation matrix

correlation_matrix = df_survey[numeric_cols].corr()

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability
plt.yticks(rotation=0)
plt.tight_layout()  # Adjust layout to fit the figure properly
plt.show()

```

```python
# Logistic Regression Plot for 'AGE' and lung cancer status
sns.lmplot(x='AGE', y='LUNG_CANCER_NUM', data=df_survey, logistic=True, ci=None)
plt.title('Logistic Regression of Age on Lung Cancer Incidence')
plt.show()
```

```python
# Box Plot for 'AGE' across lung cancer status
sns.boxplot(x='LUNG_CANCER', y='AGE', data=df_survey)
plt.title('Box Plot of Age by Lung Cancer Status')
plt.show()
```

### Trends in air quality over time and their potential impact on lung cancer rates


```python

df_air = pd.read_csv('/Users/moosarahman/Desktop/cs329e/proj/airovertime.csv')
df_air.head()

```

```python
import altair as alt

# Assuming the corrected column names are 'Date' for the year and 'AQI' for the air quality index
# Convert 'Date' to string for the selection to work properly
df_air['Date'] = df_air['Date'].astype(str)

# Create a slider for the time range
slider = alt.binding_range(min=df_air['Date'].min(), max=df_air['Date'].max(), step=1)  # Adjust the step as needed
slider_selection = alt.selection_single(bind=slider, fields=['Date'], name="Date")

# Create the line chart
line_chart = alt.Chart(df_air).mark_line(color='red').encode(
    x='Date:T',  # Use 'T' for temporal data
    y='AQI:Q',
    tooltip=['Date:T', 'AQI:Q']
).add_selection(
    slider_selection
).transform_filter(
    slider_selection
)

# Add point annotations to the line chart
text_annotations = line_chart.mark_text(
    align='left',
    baseline='middle',
    dx=7  # Nudge text to the right of the point
).encode(
    text='AQI:Q'
)

# Combine the line chart and text annotations
annotated_chart = line_chart + text_annotations

# Display the visualization
annotated_chart
```
